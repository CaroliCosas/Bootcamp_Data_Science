```python
# 1. Configuración inicial
# Librerías a utilizar:

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from collections import Counter

#2. Carga de datos
# Cargar el dataset con np.genfromtxt:

# Cargar datos
data = np.genfromtxt("madrid_idealista.csv", delimiter=",", dtype=None, encoding="utf-8", skip_header=1)

# Seleccionar columnas necesarias
# price, bedrooms, bathrooms, m2, address (quitar "Madrid")
price = data[:, 0].astype(float)
bedrooms = data[:, 1].astype(float)
bathrooms = data[:, 2].astype(float)
m2 = data[:, 3].astype(float)
addresses = np.array([addr.replace(", Madrid", "") for addr in data[:, 4]])


# 3. Análisis descriptivo (25%)

# a) Media y mediana:

# Cálculos
mean_price = np.mean(price)
median_price = np.median(price)
print(f"Media del precio: {mean_price}")
print(f"Mediana del precio: {median_price}")

# b) Máximo y mínimo:

max_price = np.max(price)
min_price = np.min(price)
print(f"Precio máximo: {max_price}")
print(f"Precio mínimo: {min_price}")

# c) Histograma y curva de densidad:

# Histograma y curva de densidad para precios
plt.figure(figsize=(10, 6))
sns.histplot(price, kde=True, bins=30, color="blue", alpha=0.6)
plt.title("Histograma y Curva de Densidad de Precios")
plt.xlabel("Precio (€)")
plt.ylabel("Frecuencia")
plt.grid()
plt.show()


# 4. Estadística avanzada (25%)

# a) Cuartiles y IQR:

# Cuartiles
q1 = np.percentile(price, 25)
q2 = np.percentile(price, 50) # Mediana
q3 = np.percentile(price, 75)
iqr = q3 - q1
print(f"Q1: {q1}, Q2: {q2}, Q3: {q3}, IQR: {iqr}")

# b) Filtrar 20% más caro y más barato:

# Filtrar 20% más barato y 20% más caro
price_sorted = np.sort(price)
top_20_percent = price_sorted[int(len(price) * 0.8):]
bottom_20_percent = price_sorted[:int(len(price) * 0.2)]

# c) Moda:

# Moda de las direcciones
address_mode = Counter(addresses).most_common(1)
print(f"Moda de las direcciones: {address_mode}")

# d) Dispersión: varianza y desviación estándar:

# Varianza y desviación estándar
variance = np.var(price)
std_dev = np.std(price)
print(f"Varianza: {variance}")
print(f"Desviación estándar: {std_dev}")


# 5. Outliers y correlación (25%)

# a) Filtro de outliers (Tukey):

# Tukey
upper_limit = q3 + 1.5 * iqr
lower_limit = q1 - 1.5 * iqr
outliers = price[(price > upper_limit) | (price < lower_limit)]
print(f"Outliers detectados: {len(outliers)}")

# b) Correlación:

# Correlación entre variables
corr_matrix = np.corrcoef([price, bedrooms, bathrooms, m2])
sns.heatmap(corr_matrix, annot=True, xticklabels=["price", "bedrooms", "bathrooms", "m2"], yticklabels=["price", "bedrooms", "bathrooms", "m2"])
plt.title("Matriz de Correlación")
plt.show()

# c) Estandarización:

# Estandarizar datos
price_std = (price - mean_price) / std_dev


# 6. Análisis de distribuciones (25%)

# a) Asimetría y curtosis:

# Asimetría y curtosis
skewness = stats.skew(price)
kurtosis = stats.kurtosis(price)
print(f"Asimetría: {skewness}")
print(f"Curtosis: {kurtosis}")

# b) Transformar distribuciones:

# Transformación logarítmica
price_log = np.log(price[price > 0]) # Evitar valores <= 0

plt.figure(figsize=(10, 6))
sns.histplot(price_log, kde=True, bins=30, color="green", alpha=0.6)
plt.title("Transformación Logarítmica de Precios")
plt.xlabel("Log(Precio)")
plt.ylabel("Frecuencia")
plt.grid()
plt.show()

# c) Contraste de hipótesis:

# Contraste 1: Casas de X barrio vs. Y barrio
barrio_x = price[addresses == "Barrio X"]
barrio_y = price[addresses == "Barrio Y"]
t_stat, p_value = stats.ttest_ind(barrio_x, barrio_y, equal_var=False)
print(f"T-stat: {t_stat}, P-value: {p_value}")

# Contraste 2: Casas con >=3 baños vs. 1-2 baños
three_bathrooms = price[bathrooms >= 3]
one_two_bathrooms = price[(bathrooms == 1) | (bathrooms == 2)]
t_stat_bath, p_value_bath = stats.ttest_ind(three_bathrooms, one_two_bathrooms, equal_var=False)
print(f"T-stat: {t_stat_bath}, P-value: {p_value_bath}")
```
