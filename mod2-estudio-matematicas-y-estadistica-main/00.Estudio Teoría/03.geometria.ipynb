## Geometría

La geometría es una rama de las matemáticas que estudia propiedades, comportamientos y relaciones entre puntos, rectas, ángulos, superficies y objetos sólidos. Es trascendental a las matemáticas, ya que muchos de estos conceptos los encontraremos en otras diferentes áreas.

- Plano Cartesiano.
- Puntos.
- Rectas.
- Distancias.
    - Tipos de distancias.

### Plano Cartesiano

El plano cartesiano son dos rectas numéricas perpendiculares, una horizontal y otra vertical, que se cortan en un punto llamado origen:

- La recta horizontal es el eje $X$ (o el eje de las abscisa).
- La recta vertical es el eje $Y$ (o el eje de las ordenadas).

### Punto
Un punto es la unidad más básica en un plano cartesiano, este es representado con números. Un punto tiene tantas dimensiones como el plano cartesiano en el que es representado. Por ejemplo, en un plano cartesiano de **2 dimensiones** los puntos son representados como $(x, y)$, donde $x$ representado su posición en el eje horizontal e $y$ representa si posición en el eje vertical.

Si hablamos de un plano de **3 dimensiones** los puntos en este plano estarán representados como $(x, y, z)$.

Si generalizamos esta definición para un plano de **n-dimensiones** los puntos en este plano estarán representados como $(x_{0}, x_{1}, x_{2}, ..., x_{n-1})$


### Recta
Una recta es una línea formada por una serie continua de puntos en una misma dirección que no tiene curvas ni ángulos y cubre la menor distancia posible entre dos puntos. En matemáticas, podemos representar una recta mediante una función o **"formula"**.

#### Ecuaciones de segundo grado:

Una ecuación de segundo grado es una ecuación polinómica cuyo grado es 2, es decir, aquella en la que el grado mayor de los monomios es 2. Puesto que la ecuación es de grado 2, tenemos, a lo sumo, 2 raíces (o soluciones) distintas. La grafica de una ecuación de segundo grado es siempre una curva.

$$x^2 -x -2 = 0$$

nos sirve para hacer predicciones, por ejemplo sobre un scatter de puntos y podemos hacer uns regresion lineal sobre ellos


```python
x = np.linspace(-5, 5, 100)
y = x**2 - x - 2

# Curva
plt.plot(x, y)

# Eje X
plt.axhline(0, color = "black")

# Eje Y
plt.axvline(0, color = "black")

plt.grid()
plt.show()
```

### Distancia Euclideana

Si tenemos los puntos **P1** y **P2**:

$$P1 = (1, 2) = (x_1, y_1)$$
$$P2 = (4, 5) = (x_2, y_2)$$

La distancia entre dos puntos esta dada por la siguiente ecuación:

$$D(P1,P2) = \sqrt[]{(x_2-x_1)^2 + (y_2-y_1)^2} $$

$$D(P1,P2) = \sqrt[]{(4-1)^2+(5-2)^2}$$

$$D(P1,P2) = \sqrt[]{9+9} = \sqrt[]{18}$$

$$ D(P1,P2) = 4.242640687119285$$

La distancia entre dos puntos en un espacio de tres dimensiones esta dada por la siguiente ecuación: 

$$D(P1,P2) = \sqrt[]{(x_2-x_1)^2 + (y_2-y_1)^2 + (z_2-z_1)^2} $$

Si seguimos aumentando dimensiones, se agregan a la ecuación para encontrar la distancia entre ambos puntos dependiendo de las dimensiones en las que se encuentren.

$$ P1 = (x_1,y_1,z_1,k_1, ..., n_1) $$
$$ P2 = (x_2,y_2,z_2,k_2, ..., n_2) $$

$$D(P1, P2) = \sqrt[]{(x_2-x_1)^2 + (y_2-y_1)^2 + (z_2-z_1)^2 + (k_2-k_1)^2 +...+ (n_2-n_1)^2} $$

$$D(P1, P2) = \sqrt[]{\sum_{i = 0}^{n} (x_{i} - y_{i})^{2}}$$

A esta forma de encontrar la distancia entre dos puntos la podemos llamar: **distancia euclidiana**

Existen diferentes formas de calcular la distancia entre 2 puntos:

### Distancia Manhattan

La distancia Manhattan es una medida de la distancia entre dos puntos en un espacio vectorial. Se calcula como la suma de las diferencias absolutas de las coordenadas de los puntos.

Se puede visualizar como la distancia que recorre un taxista al moverse por una cuadrícula de calles, donde solo puede girar en ángulos rectos. (ya que no se pueden atravesar los edificios)

$$\text {Manhattan} = \sum_{i = 0}^{n}\left| x_{i} - y_{i} \right|$$

### Distancia de Chebyshev

La distancia de **Chebyshev** es la mayor de las diferencias entre las dimensiones de dos puntos.

Se puede visualizar como la distancia máxima que se debe recorrer en cualquier dirección para pasar de un punto a otro.

$$\text{Chebysev} = max(\left| x_{1} - x_{0} \right|, \left| y_{1} - y_{0} \right|, ..., \left| k_{1} - k_{0} \right|)$$

### Distancia Minkowski

La distancia Minkowski es una medida de la distancia entre dos puntos en un espacio vectorial.

Es una generalización de la distancia euclidiana y la distancia de Manhattan, y se puede utilizar para medir la distancia entre puntos en cualquier espacio vectorial con norma.

$$
\text{Minkowski} = \left(\sum_{i=1}^{n}\left|x_i - y_i\right|^{p}\right)^{\frac{1}{p}}
$$

Cuando $p = 1$ obtenemos la distancia Manhattan:

$$\left(\sum_{i=1}^{n}|x_i-y_i|^{1}\right)^{\frac{1}{1}} = \sum_{i = 0}^{n}\left| x_{i} - y_{i} \right|$$


Cuando $p = 2$ obtenemos la distancia euclideana:

$$\left(\sum_{i=1}^{n}|x_i-y_i|^{2}\right)^{\frac{1}{2}} = \sqrt[]{\sum_{i = 0}^{n} (x_{i} - y_{i})^{2}}$$
